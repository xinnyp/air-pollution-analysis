```python
df_lambeth_2015 = Lambeth.get_pollution_data(22,2015)
df_lambeth_2014 = Lambeth.get_pollution_data(22,2014)
df_list = [df_lambeth_2014, df_lambeth_2015, df_lambeth_2016, df_lambeth_2017, df_lambeth_2018]
```

```python
df_pm10 = pd.DataFrame()
for df in df_list:
    df_pm10 = df_pm10.append(df.loc[(df['species_code'] == 'PM10') & (df['objective_name']=='Capture Rate (%)')])
df_pm10
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df_pm10['value'] = df_pm10['value'].apply(pd.to_numeric)
plt.figure(figsize=(16, 6))
g = sns.lineplot(x="year", y="value", hue="site_name", data=df_pm10, marker="o")
```
