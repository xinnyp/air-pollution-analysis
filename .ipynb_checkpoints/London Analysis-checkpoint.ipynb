### Instructions for use
-In order to fetch data for a borough/London, create a GetData object

-The object has the following methods:
  1. get_local_authority_data(get_id_only=False), this will return the full JSON object from API. set get_id_only to True to return id only
  2. get_pollution_data(authority_id:int, year:List), this will return average yearly pollution data based on the year provided.
  3. get_daily_pollution_stats(local_authority_ids:List), this will return today's pollution stats for all sites at the borough selected.

```python
hackney = GetData('Hackney')
lambeth = GetData('lambeth')
westminster = GetData('westminster')
london = GetData('london')
hillingdon = GetData('hillingdon')
croydon = GetData('croydon')
ealing = GetData('ealing')
kc = GetData('Kensington and Chelsea')
```

```python
locations_list = [hackney, lambeth, westminster, hillingdon, croydon, ealing]
```

```python
id_list = []
for i in locations_list:
    authority_id = i.get_local_authority_data(get_id_only= True)
    id_list.append(authority_id)
print(id_list)
```

```python
london.get_pollution_data(17, [2018])
```

```python
london_pollution_df = london.get_daily_pollution_stats(id_list)
```

```python
# multiple sites may be present in one borough so group and calc average air_index_quality
london_pollution_pm10_df = london_pollution_df.loc[london_pollution_df['species_code'] == 'PM10']
london_pollution_pm10_df=london_pollution_pm10_df.groupby('borough')['air_quality_index'].mean().to_frame('average_air_quality_index').reset_index()
london_pollution_pm10_df
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# set the filepath and load in a shapefile

fp = '/Users/Xinny/PycharmProjects/AirPollutionAnalysis/geodata/London_Borough_Excluding_MHW.shp'

map_df = gpd.read_file(fp)

merged = map_df.set_index('NAME').join(london_pollution_pm10_df.set_index('borough'))
```

```python
visualise_var = 'average_air_quality_index'

# first layer, all geometries included 
base = merged.plot(color="grey",figsize=(10,6))

# second layer, NaN geometries excluded
merged.dropna(subset=['average_air_quality_index']).plot(column=visualise_var, ax=base, legend=True, cmap='OrRd')

# styling plot 
base.axis('off')
base.set_title('air_quality_index', fontdict={'fontsize':'25', 'fontweight':3})
base.annotate('Source: KCL London Air API, 2019', xy=(0.1, 0.08), xycoords='figure fraction')
```
