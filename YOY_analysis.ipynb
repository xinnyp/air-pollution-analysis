```python
from data_fetcher.get_data import GetData
import pandas as pd
```

```python
Lambeth = GetData('Lambeth')
Hackney = GetData('hackney')
```

```python
df_lambeth = Lambeth.get_pollution_data(22, [2018, 2017, 2016, 2014, 2015])
df_lambeth.head()
```

```python
df_pm10 = df_lambeth.loc[(df_lambeth['species_code'] == 'PM10') & (df_lambeth['objective_name']=='Capture Rate (%)')]
df_pm10
```

```python
df_hackney = Hackney.get_pollution_data(12,[2014,2015,2016,2017,2018])
df_hackney.head()
```

```python
Hackney.plot_pollution_by_species('PM10', df_hackney)
```

```python
Hillingdon = GetData('Hillingdon')
df_hillingdon = Hillingdon.get_pollution_data(17,[2014,2015,2016,2017,2018])
df_hillingdon.head()
```

```python
Hillingdon.plot_pollution_by_species('O3',df_hillingdon)
```

```python
df_combined_councils = pd.concat([df_hackney, df_lambeth, df_hillingdon])
Lambeth.plot_pollution_by_species('PM10', df_combined_councils)
```
